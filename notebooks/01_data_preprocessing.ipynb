# Data Preprocessing
## Dataset Overview

```python
# Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])
df.isnull().sum()
```

```python
# Remove Cancelled Invoices
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

```python
# Remove negative Quantities
df[df['Quantity'] < 0].shape
df = df[df['Quantity'] > 0]
df[df['Quantity'] <= 0].shape
```

```python
# Convert InvoiceDate into Datetime Format
df['InvoiceDate'].dtype
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceDate'].dtype
```

```python
# Create TotalPrice Column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
# Check for negative quantities
negative_quantities = df[df['Quantity'] < 0]
negative_quantities.shape
```

```python
# Check Dataset Shape After Cleaning
df.shape
```

```python
df.to_csv("../data/cleaned_online_retail.csv", index=False)
```
